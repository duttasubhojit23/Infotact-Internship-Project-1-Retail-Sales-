```python

import pandas as pd
import matplotlib.pyplot as plt

file = "DAproject.xlsx"
```

# Retail Sales EDA
This notebook performs exploratory data analysis on retail sales data to uncover patterns and insights.

```python
# ----------- FUNCTION TO CLEAN -----------
def clean_sheet(sheet):
    df = pd.read_excel(file, sheet_name=sheet, header=None)
    header_row = df.notnull().sum(axis=1).idxmax()
    df.columns = df.iloc[header_row]
    df = df[header_row + 1:]
    df = df.dropna(axis=1, how='all').dropna(how='all').reset_index(drop=True)
    df.columns = df.columns.str.strip().str.replace(" ", "_")
    return df
```

## Data Cleaning
This step ensures the dataset is consistent by handling missing values and duplicates.

## Data Exploration
Understanding distribution and relationships between variables.

```python
# ================= PRODUCT CATEGORY =================
product = clean_sheet("Product_Category")
product["Total_Revenue"] = pd.to_numeric(product["Total_Revenue"], errors='coerce')
product["Total_Quantity"] = pd.to_numeric(product["Total_Quantity"], errors='coerce')

print("Product EDA:", product.shape)
print(product.isnull().sum())

product_summary = product.groupby(["Category","Product_Id"])["Total_Revenue"].sum().reset_index()
product_summary.to_csv("product_performance.csv", index=False)

pivot = product_summary.pivot(index="Product_Id", columns="Category", values="Total_Revenue")
pivot.plot(kind="bar", figsize=(10,5))
plt.title("Product Revenue by Category")
plt.xticks(rotation=45)
plt.show()

```

```python
# ================= CITY CATEGORY =================
city = clean_sheet("City_Category")
city["Revenue"] = pd.to_numeric(city["Revenue"], errors='coerce')

print("City EDA:", city.shape)
print(city.isnull().sum())

city_summary = city.groupby(["City","Category"])["Revenue"].sum().reset_index()
city_summary.to_csv("city_category_sales.csv", index=False)

pivot = city_summary.pivot(index="Category", columns="City", values="Revenue")
pivot.plot(kind="bar", figsize=(10,5))
plt.title("Category Revenue by City")
plt.xticks(rotation=45)
plt.show()
```

```python
# ================= MONTHLY CATEGORY =================
monthly = clean_sheet("Monthly_Category")
monthly["Revenue"] = pd.to_numeric(monthly["Revenue"], errors='coerce')
monthly["Month"] = pd.to_datetime(monthly["Month"])

print("Monthly EDA:", monthly.shape)
print(monthly.isnull().sum())

monthly_summary = monthly.groupby(["Month","Category"])["Revenue"].sum().reset_index()
monthly_summary.to_csv("monthly_category_sales.csv", index=False)

pivot = monthly_summary.pivot(index="Month", columns="Category", values="Revenue")

fig, ax = plt.subplots(figsize=(10,5))
pivot.plot(kind="bar", ax=ax)
pivot.sum(axis=1).plot(ax=ax, marker='o')
plt.title("Monthly Revenue Trend")
plt.xticks(rotation=45)
plt.show()
```

```python
# ================= HOURLY SALES =================
hourly = clean_sheet("Hourly_Sales")
hourly["Hour"] = pd.to_numeric(hourly["Hour"], errors='coerce')
hourly["Total_Orders"] = pd.to_numeric(hourly["Total_Orders"], errors='coerce')

print("Hourly EDA:", hourly.shape)
print(hourly.isnull().sum())

hourly = hourly.sort_values("Hour")
hourly.to_csv("hourly_orders.csv", index=False)

fig, ax = plt.subplots(figsize=(10,5))
ax.bar(hourly["Hour"], hourly["Total_Orders"])
ax.plot(hourly["Hour"], hourly["Total_Orders"], marker='o')
plt.title("Hourly Orders")
plt.show()
```

```python
# ================= DAILY SALES =================
daily = clean_sheet("Daily_Sales")
daily["Revenue"] = pd.to_numeric(daily["Revenue"], errors='coerce')
daily["Day"] = pd.to_datetime(daily["Day"])

print("Daily EDA:", daily.shape)
print(daily.isnull().sum())

daily = daily.sort_values("Day")
daily.to_csv("daily_sales.csv", index=False)

plt.figure(figsize=(12,5))
plt.plot(daily["Day"], daily["Revenue"], marker='*')
plt.title("Daily Revenue Trend")
plt.xticks(rotation=45)
plt.show()
```

```python
# ================= SEGMENT CATEGORY =================
segment = clean_sheet("Segment_Category")
segment["Revenue"] = pd.to_numeric(segment["Revenue"], errors='coerce')

print("Segment EDA:", segment.shape)
print(segment.isnull().sum())

segment_summary = segment.groupby(["Customer_Segment","Category"])["Revenue"].sum().reset_index()
segment_summary.to_csv("segment_category_sales.csv", index=False)

pivot = segment_summary.pivot(index="Category", columns="Customer_Segment", values="Revenue")
pivot.plot(kind="bar", figsize=(10,5))
plt.title("Segment vs Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# ================= TOP PRODUCT CITY =================
top_city = clean_sheet("Top_Product_City")
top_city["Revenue"] = pd.to_numeric(top_city["Revenue"], errors='coerce')

print("Top City EDA:", top_city.shape)
print(top_city.isnull().sum())

top_product_city = top_city.groupby(["City","Product_Id"])["Revenue"].sum().reset_index()
top_product_city.to_csv("top_product_city.csv", index=False)

pivot = top_product_city.pivot(index="City", columns="Product_Id", values="Revenue")
pivot.plot(kind="bar", figsize=(12,5))
plt.title("Product Revenue by City")
plt.xticks(rotation=45)
plt.show()
```

```python
# ================= TOP PRODUCT CITY =================
top_city = clean_sheet("Top_Product_City")
top_city["Revenue"] = pd.to_numeric(top_city["Revenue"], errors='coerce')

print("Top City EDA:", top_city.shape)
print(top_city.isnull().sum())

top_product_city = top_city.groupby(["City","Product_Id"])["Revenue"].sum().reset_index()
top_product_city.to_csv("top_product_city.csv", index=False)

pivot = top_product_city.pivot(index="Product_Id", columns="City", values="Revenue")
pivot.plot(kind="bar", figsize=(12,5))
plt.title("Product Revenue by City")
plt.xticks(rotation=45)
plt.show()
```

```python
# ================= KPI =================
kpi = clean_sheet("KPI_VALUES")
kpi.columns = kpi.columns.str.strip()

kpi["Total_Revenue"] = pd.to_numeric(kpi["Total_Revenue"], errors='coerce')
kpi["Total_Orders"] = pd.to_numeric(kpi["Total_Orders"], errors='coerce')

total_revenue = kpi["Total_Revenue"].sum()
total_orders = kpi["Total_Orders"].sum()
avg_order_value = total_revenue / total_orders

kpi_summary = pd.DataFrame({
    "Metric": ["Total Revenue", "Total Orders", "Avg Order Value"],
    "Value": [total_revenue, total_orders, avg_order_value]
})

kpi_summary.to_csv("kpi_summary.csv", index=False)

kpi_summary.set_index("Metric").plot(kind="bar", figsize=(6,4))
plt.title("KPI Overview")
plt.show()
```
